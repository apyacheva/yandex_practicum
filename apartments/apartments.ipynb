# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. **Наша задача** — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание полей данных:**\
`airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)\
`balcony` — число балконов\
`ceiling_height` — высота потолков (м)\
`cityCenters_nearest` — расстояние до центра города (м)\
`days_exposition` — сколько дней было размещено объявление (от публикации до снятия)\
`first_day_exposition` — дата публикации\
`floor` — этаж\
`floors_total` — всего этажей в доме\
`is_apartment` — апартаменты (булев тип)\
`kitchen_area` — площадь кухни в квадратных метрах (м²)\
`last_price` — цена на момент снятия с публикации\
`living_area` — жилая площадь в квадратных метрах (м²)\
`locality_name` — название населённого пункта\
`open_plan` — свободная планировка (булев тип)\
`parks_around3000` — число парков в радиусе 3 км\
`parks_nearest` — расстояние до ближайшего парка (м)\
`ponds_around3000` — число водоёмов в радиусе 3 км\
`ponds_nearest` — расстояние до ближайшего водоёма (м)\
`rooms` — число комнат\
`studio` — квартира-студия (булев тип)\
`total_area` — площадь квартиры в квадратных метрах (м²)\
`total_images` — число фотографий квартиры в объявлении

## Предобработка данных

Для начала проверим, есть ли в наших данных дублирующиеся строки.

```python
data.duplicated().sum()
```

Судя по всему, явных дубликатов нет.

Будем рассматривать данные в каждом столбце по очереди и сразу принимать решения по преобразованию данных.

###### `locality_name`

Начнем с населенного пункта. Выведем список уникальных значений.

```python
data['locality_name'].sort_values().unique()
```

Значений очень много, глазами проверить сложно, но, по крайней мере, бросается в глаза тот факт, что в одних и тех же словах встречается то `е`, то `ё`. Заменим все буквы `ё` на `е`.

```python
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
```

```python
data['locality_name'].sort_values().unique()
```

Снова проверим на дубликаты.

```python
data.duplicated().sum()
```

Хорошо, что они не появились.
В столбце есть пропуски, всего 49 шт, мало, но есть. Поскольку мы не можем знать, к какому населенному пункту должны относиться эти записи, заменим пропущенные значения на `unknown`

```python
data['locality_name'] = data['locality_name'].fillna('unknown')
# проверим
data['locality_name'].isna().sum()
```

Общее количество этажей сложно откуда-то взять. Единственное, можно с уверенностью сказать, что оно точно не меньше этажа, на котором расположена квартира. Но не будем заменять пропуски значением текущего этажа, т.к. в этом случае все эти данные при дальнейшем анализе влияния этажа на стоимость квартиры попадут в категорию "последний этаж" и исказят результат. Лучше сделать так, чтобы они попали в "другой этаж", т.к. таких данных очевидно больше и влияние на результат окажется меньше. Поэтому заполним пропуски в столбце этажности дома значением "текущий этаж+1"

```python
data.loc[data['floors_total'].isna(), 'floors_total'] = data['floor'] + 1
# проверим пропуски 
data[data['floors_total'].isna()]['floors_total'].sum()
```

"Квартира" с высотой потолков 100 м имеет общую площадь 25 кв.м., 1 комнату и при этом 5 балконов. Как будто бы тут посчитан целый подъезд, но даже подъезд пятиэтажного дома не будет иметь выысоту 100 м. Логичнее всего будет удалить эту строку.\
Высота потолков от 20 до 32 очень похожа на опечатку в данных. Возможно вместо метров указаны дециметры. Т.е. если мы разделим эти значения на 10, в результате получим адекватные данные.\
Но сначала заполним пропуски в этом столбце. Пропуски в числовых данных логичнее всего заполнять медианным значением, чтобы нивелировать влияние выбросов (очень больших или очень маленьких значений). В данном случае медиану лучше взять по группам, т.к. высота потолков может различаться в домах разной этажности (построенных по разным проектам), а так же в разных населенных пунктах. Создадим для этого новый столбец.

```python
# удаляем строку с 100м
data = data.drop(data[data['ceiling_height'] == 100].index)
# заполняем пропуски медианными значениями по группам
data['ceiling_height_new'] = data['ceiling_height'].fillna(data.groupby(['locality_name', 'floors_total'])['ceiling_height'].transform('median'))
# делим слишком большие значения на 10
data.loc[data['ceiling_height'] >= 20, 'ceiling_height_new'] = data['ceiling_height'] / 10
# проверим
data[data['ceiling_height'] >= 20]
```

Значения нового столбца `ceiling_height_new` корректные. Перезапишем эти значения в первоначальный столбец `ceiling_height`

```python
data['ceiling_height'] = data['ceiling_height_new']
data = data.drop(['ceiling_height_new'], axis=1)
```

```python
data['ceiling_height'].isna().sum()
```

В столбце однако остались еще пропуски, вероятно, это связано с невозможностью посчитать медиану по группам. Ограничимся теперь группировкой по одному параметру - этажность здания.

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby('floors_total')['ceiling_height'].transform('median'))    
data[data['ceiling_height'].isna()]
```

Осталась всего одна строка, т.к. 33-этажное здание встречается в данных всего 1 раз. Заполним его медианным значением по Санкт-Петербургу без учета этажности. Даже если это будет не очень правильно, одно значение на результат сильно не повлияет.

###### `living_area`

В этом столбце есть  1903 пропуска. Выведем на экран часть строк с пропущенными значениями в этом столбце.

```python
data[data['living_area'].isna()].head(5)
```

Строки с пропущенными значениями жилой площади также имеют пропуски в площади кухни, значит никак не посчитать одно через другое.\
Очевидно, что жилая площадь зависит от общей. Разобьем общую площадь на группы и посчитаем жил.площадь как медиану соответствующих жилых площадей для каждой группы.\
Создадим следующие группы:\
1 - общая площадь до 30 кв.м\
2 - от 31 до 40\
3 - от 41 до 50\
4 - от 51 до 60\
5 - от 61 до 70\
6 - от 71 до 80\
7 - от 81 до 100\
8 - свыше 100

```python
def area_group(area):
    if area <= 30:
        return 1
    elif area <= 40:
        return 2
    elif area <= 50:
        return 3
    elif area <= 60:
        return 4
    elif area <= 70:
        return 5
    elif area <= 80:
        return 6
    elif area <= 100:
        return 7
    else:
        return 8

# создадим новый столбец для группировки по площади    
data['total_area_group'] = data['total_area'].apply(area_group)
data.head()
```

Теперь на основе столбца с группами посчитаем медиану и заполним пропуски

```python
data['living_area'] = data['living_area'].fillna(data.groupby('total_area_group')['living_area'].transform('median'))
data['living_area'].isna().sum()
```

Для удобства дальнейшего анализа разобьем все квартиры на категории в зависимости от этажа, на котором они расположены: первый, последний, другой.\
Для этого напишем функцию, которая будет построчно обрабатывать дата-фрейм.

```python
def floor(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'

# вызовем функцию
data['floor_category'] = data.apply(floor, axis=1)
data.head()
```
